#### Calculating great circle distance

```python
def haversine_np(lon1, lat1, lon2, lat2):
    """
    Calculate the great circle distance between two points
    on the earth (specified in decimal degrees)
    lon1, lat1, lon2, lat2: can all take scalar value or vector value
    All args must be of equal length.    

    """
    lon1, lat1, lon2, lat2 = map(np.radians, [lon1, lat1, lon2, lat2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = np.sin(dlat/2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2.0)**2

    c = 2 * np.arcsin(np.sqrt(a))
    miles = 6367 * c *0.621371
    
    return miles
```

```python
#Approach I: apply the great_circle() function in geopy to each row of df
from geopy.distance import great_circle
df['distance'] = df.apply(lambda x: great_circle((x['lat1'], x['lon1']), (x['lat2'], x['lon2'])).miles, axis=1)

#Approach II: use the above vectorized numpy version
df['distance'] = haversine_np(df['lon1'], df['lat1'], df['lon2'], df['lat2'])

#Performances: for a df with 4.6 million rows
#Approach I: 378.84809494 seconds
#Approach II: 0.524402141571 seconds
#Difference: around 0.06% (average diff/average distance)
```

```python
#define treatment and control group
df_t, df_c = df[(df['SQ']>0)], df[(df['SQ']==0)] #t means treatment, c means control
treatment, control = 'SQ', 'no SQ'
inputs = [df_t, df_c, treatment, control]
df_t.shape[0], df_c.shape[0]
```

```python
metrics = ['posOutcome','mutualEngagement','posOutcomeOrME','qualifiedApply']
utility.mean_std_for_2_groups(*(inputs+[metrics]))
```

```python
print("histogram comparison")
utility.hist_compare(df_t, df_c, treatment, control, metrics, save_path = "graph/"+"histogram"+"_mutualEngagement")

# print("empirical cdfs")
# utility.ecdf_compare(df_t, df_c, treatment, control, metrics, save_path = "graph/"+"ecdf"+"_mutualEngagement")
```

### Hypothesis testing

```python
#Hypotheses: H0: mu_t=mu_c vs. H1: mu_t>mu_c (since SQ is expected to help on mutual engagement related metrics)
for metric in metrics:
    t = utility.t_test(df_t[metric], df_c[metric])
    p = utility.p_value(t, 't>c')
    print("t statistic ({}): {}, p value: {}".format(metric, t, p))
```
